```python
import numpy as np
from scipy import stats
# Nivel de significancia
alpha = 0.05


# Asumiendo que 'count' representa el tamao de la muestra para cada fila
df10['df'] = df10['count'] - 1  # Grados de libertad
df10['t - critico'] = stats.t.ppf(1 - alpha/2, df10['df']) # Grados de libertad

# Aplicamos la frmula del IC con t de Student
df10['IC()_t - 95%'] = df10.apply(lambda row: (
    round(row['mean'] - stats.t.ppf(1 - alpha/2, row['df']) * row['std'] / np.sqrt(row['count']), 2),
    round(row['mean'] + stats.t.ppf(1 - alpha/2, row['df']) * row['std'] / np.sqrt(row['count']), 2)
), axis=1)

df10
```

### **b. Intervalo de confianza con $ = 0.02$ - Es decir, 98% de confianza**

```python
import numpy as np
from scipy import stats
# Nivel de significancia
alpha = 0.02


# Asumiendo que 'count' representa el tamao de la muestra para cada fila
df10['df'] = df10['count'] - 1  # Grados de libertad
df10['t - critico'] = stats.t.ppf(1 - alpha/2, df10['df']) # Grados de libertad

# Aplicamos la frmula del IC con t de Student
df10['IC()_t - 98%'] = df10.apply(lambda row: (
    round(row['mean'] - stats.t.ppf(1 - alpha/2, row['df']) * row['std'] / np.sqrt(row['count']), 2),
    round(row['mean'] + stats.t.ppf(1 - alpha/2, row['df']) * row['std'] / np.sqrt(row['count']), 2)
), axis=1)

df10
```
