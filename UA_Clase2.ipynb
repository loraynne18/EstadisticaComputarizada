```python
# prompt: Crea una tabla de frecuencia y el histograma con su poligono de frecuencia con la variable NRC

import matplotlib.pyplot as plt
# @title **Paso 10. Tabla de frecuencia y histograma para la variable NRC**

# Tabla de frecuencia
tabla_frecuencia_nrc = df['NRC'].value_counts().reset_index()
tabla_frecuencia_nrc.columns = ['NRC', 'Frecuencia Absoluta']
tabla_frecuencia_nrc['Frecuencia Relativa'] = tabla_frecuencia_nrc['Frecuencia Absoluta'] / len(df)
tabla_frecuencia_nrc['Frecuencia Acumulada'] = tabla_frecuencia_nrc['Frecuencia Absoluta'].cumsum()
tabla_frecuencia_nrc['Frecuencia Relativa Acumulada'] = tabla_frecuencia_nrc['Frecuencia Relativa'].cumsum()

print("Tabla de Frecuencia para NRC:")
print(tabla_frecuencia_nrc)

# Histograma y polígono de frecuencia
plt.figure(figsize=(15, 7))
sns.histplot(data=df, x='NRC', kde=False, color='skyblue', bins=len(df['NRC'].unique()))
plt.title('Histograma y Polígono de Frecuencia de NRC')
plt.xlabel('NRC')
plt.ylabel('Frecuencia')

# Para el polígono de frecuencia, necesitamos los puntos medios de las barras del histograma
# y las frecuencias de cada NRC. Como NRC es categórica, los puntos medios son los NRCs mismos.
# Las frecuencias son las de la tabla de frecuencia.
nrc_counts = df['NRC'].value_counts().sort_index()
nrc_values = nrc_counts.index
nrc_frequencies = nrc_counts.values

# Añadir los puntos inicial y final para que el polígono toque el eje x
nrc_values_poly = [nrc_values[0] - 1] + list(nrc_values) + [nrc_values[-1] + 1]
nrc_frequencies_poly = [0] + list(nrc_frequencies) + [0]

plt.plot(nrc_values_poly, nrc_frequencies_poly, marker='o', linestyle='-', color='red', label='Polígono de Frecuencia')
plt.xticks(rotation=90)
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

```python
# prompt: Crea una tabla de frecuencia y el histograma con su poligono de frecuencia con la variable PESO

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Crear la tabla de frecuencia para la variable PESO
# Determinar el número de clases (regla de Sturges)
k = 1 + 3.322 * np.log10(len(df))
k = int(round(k))

# Calcular el rango
rango = df['PESO'].max() - df['PESO'].min()

# Calcular el ancho de clase
ancho_clase = rango / k

# Definir los límites de clase
bins = np.arange(df['PESO'].min(), df['PESO'].max() + ancho_clase, ancho_clase)
labels = [f'[{bins[i]:.2f} - {bins[i+1]:.2f})' for i in range(len(bins)-2)] + [f'[{bins[-2]:.2f} - {bins[-1]:.2f}]']

df['Intervalo_Peso'] = pd.cut(df['PESO'], bins=bins, right=False, labels=labels, include_lowest=True)

tabla_frecuencia_peso = df['Intervalo_Peso'].value_counts().sort_index().reset_index()
tabla_frecuencia_peso.columns = ['Intervalo de Peso', 'Frecuencia Absoluta']

# Calcular frecuencias relativas y acumuladas
tabla_frecuencia_peso['Frecuencia Relativa'] = tabla_frecuencia_peso['Frecuencia Absoluta'] / len(df)
tabla_frecuencia_peso['Frecuencia Acumulada'] = tabla_frecuencia_peso['Frecuencia Absoluta'].cumsum()
tabla_frecuencia_peso['Frecuencia Relativa Acumulada'] = tabla_frecuencia_peso['Frecuencia Relativa'].cumsum()

print("\nTabla de Frecuencia para PESO:")
print(tabla_frecuencia_peso)

# Crear el histograma y polígono de frecuencia para la variable PESO
plt.figure(figsize=(10, 6))

# Histograma
hist, bin_edges = np.histogram(df['PESO'], bins=bins)
sns.histplot(data=df, x='PESO', bins=bins, kde=False, color='skyblue', alpha=0.7, label='Histograma')

# Polígono de frecuencia
# Calcular los puntos medios de los intervalos
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

# Añadir puntos inicial y final para que el polígono toque el eje x
bin_centers_poly = np.insert(bin_centers, 0, bin_centers[0] - ancho_clase)
bin_centers_poly = np.append(bin_centers_poly, bin_centers[-1] + ancho_clase)

hist_poly = np.insert(hist, 0, 0)
hist_poly = np.append(hist_poly, 0)

plt.plot(bin_centers_poly, hist_poly, marker='o', linestyle='-', color='red', label='Polígono de Frecuencia')

plt.title('Histograma y Polígono de Frecuencia de PESO')
plt.xlabel('PESO')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()
```
